```python
#assigning the values
predict="price"
data=df[["symboling", "wheelbase", "carlength", "carwidth", "carheight", "curbweight", "enginesize", "boreratio", "stroke", "compressionratio", "horsepower", 
             "peakrpm", "citympg", "highwaympg", "price"]]
x=np.array(data.drop([predict],1))
y=np.array(data[predict])
```

```python
#spliting of dataset into train and test sets for training and testing the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#shapes of the sets
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#we use DECISION TREE REGRESSOR to measure the quality of a split.
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
```

```python
#Traning the model
model.fit(x_train,y_train)
```

```python
#prediction
y_pred=model.predict(x_test)
y_pred
```

```python
#model evaluation
from sklearn.metrics import mean_absolute_error
model.score(x_test,y_pred)
```
